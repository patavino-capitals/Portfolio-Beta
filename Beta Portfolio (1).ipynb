```python
import yfinance as yf
import pandas as pd

tickers = ['TRV', 'AIG', 'PGR', 'CB', 'MMC', 'AON', 'AJG', 'AFL', 'MFC.TO', 'ALL']

start_date = "2014-11-03"
end_date = "2024-11-03"

market_caps = pd.DataFrame()

for ticker in tickers:
    print(f"Fetching data for {ticker}...")
    stock = yf.Ticker(ticker)
    
    # Download historical data
    hist_data = stock.history(start=start_date, end=end_date, interval="1mo")
    
    # Get market cap from close price and shares outstanding
    if 'Close' in hist_data.columns:
        hist_data['Market Cap'] = hist_data['Close'] * stock.info.get('sharesOutstanding', float('nan'))
        market_caps[ticker] = hist_data['Market Cap']

market_caps
```
